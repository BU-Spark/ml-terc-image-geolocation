### Labeled location vs predicted location graphs

```python
import pandas as pd
from geopy.distance import geodesic
import numpy as np
import matplotlib.pyplot as plt
import requests
from PIL import Image
from io import BytesIO
from IPython.display import display, clear_output
import os
```

```python
# Function to create a Google Maps image URL for satellite images with markers
def create_google_maps_url(api_key, lat, lng, zoom=9):
    # Marker format: size:mid|color:red|label:X|lat,lng
    marker = f"size:mid|color:red|{lat},{lng}"
    return f"https://maps.googleapis.com/maps/api/staticmap?center={lat},{lng}&zoom={zoom}&size=400x400&maptype=satellite&markers={marker}&key={api_key}"

# Function to display images side by side
def display_side_by_side(*images):
    total_width = sum(i.size[0] for i in images)
    max_height = max(i.size[1] for i in images)
    new_im = Image.new('RGB', (total_width, max_height))

    x_offset = 0
    for im in images:
        new_im.paste(im, (x_offset,0))
        x_offset += im.size[0]

    display(new_im)

# Enter api_key
api_key = input("Enter Google Maps API Key")
# Clear the console
clear_output(wait=True)
# Read the CSV file
df = pd.read_csv('../results/results_SIFT_v2_csv.csv')

# Add a new column for minimum distance
df['Min Distance'] = df.apply(lambda row: min(row['Distance Labeled to Predicted 1'], row['Distance Labeled to Predicted 2']), axis=1)

# Sort the DataFrame by the 'Min Distance' column
df_sorted = df.sort_values(by='Min Distance')

# Process each row to generate and display maps
for index, row in df_sorted.iterrows():
    # Extract the unique identifier and locations
    identifier = row.iloc[0]  # Replace with the correct column index if needed
    loc = row['Location'].split(', ')
    pred_loc1 = row['Predicted Location 1'].split(', ')
    pred_loc2 = row['Predicted Location 2'].split(', ')
    dist1 = row['Distance Labeled to Predicted 1']
    dist2 = row['Distance Labeled to Predicted 2']

    # Determine which predicted location to use
    pred_loc = pred_loc1 if dist1 < dist2 else pred_loc2

    # Print image names or descriptions
    print(f"{identifier}: Labeled Location ({loc[0]}, {loc[1]}) vs Predicted Location ({pred_loc[0]}, {pred_loc[1]})")

    # Create URLs for Google Maps images
    loc_url = create_google_maps_url(api_key, loc[0], loc[1])
    pred_loc_url = create_google_maps_url(api_key, pred_loc[0], pred_loc[1])

    # Get images from URLs
    loc_img = Image.open(BytesIO(requests.get(loc_url).content))
    pred_loc_img = Image.open(BytesIO(requests.get(pred_loc_url).content))

    # Display images side by side
    display_side_by_side(loc_img, pred_loc_img)

    # You may remove the break statement to process all rows
    # break
```

### Error rate analysis

```python
df = pd.read_csv('../results/results_SIFT_v2_csv.csv')
```

```python
# Sort the DataFrame based on two columns
df_sorted = df.sort_values(by=["Distance Labeled to Predicted 1", "Distance Labeled to Predicted 2"])

# Display the first few rows of the sorted DataFrame
df_sorted.head()
```

```python
def calculate_distance(coord1, coord2):
    """
    Calculate distance between two coordinates
    """
    try:
        return geodesic(coord1, coord2).km
    except:
        return np.nan

def extract_coordinates(coord_string):
    """
    Extract coordinates from a string
    """
    if pd.isna(coord_string):
        return np.nan, np.nan
    coords = coord_string.split(", ")
    return float(coords[0]), float(coords[1])

# Extract coordinates for actual location, best match 1, best match 2, and predicted location
df['Location_coords'] = df['Location'].apply(lambda x: extract_coordinates(x))
df['Best_Match_1_coords'] = df['Predicted Location 1'].apply(lambda x: extract_coordinates(x))
df['Best_Match_2_coords'] = df['Predicted Location 2'].apply(lambda x: extract_coordinates(x))

# Calculate distances
df['Error_Best_Match_1'] = df.apply(lambda x: calculate_distance(x['Location_coords'], x['Best_Match_1_coords']), axis=1)
df['Error_Best_Match_2'] = df.apply(lambda x: calculate_distance(x['Location_coords'], x['Best_Match_2_coords']), axis=1)

df
```

```python
# Calculate error rates
error_rate_best_match_1 = df['Error_Best_Match_1'].mean()
error_rate_best_match_2 = df['Error_Best_Match_2'].mean()

print("Error rate for best match 1: {:.2f} km".format(error_rate_best_match_1))
print("Error rate for best match 2: {:.2f} km".format(error_rate_best_match_2))
```

```python
def calculate_lat_long_error(coord1, coord2):
    """
    Calculate the difference in latitude and longitude between two coordinates.
    """
    lat1, lon1 = coord1
    lat2, lon2 = coord2
    
    # Calculate differences
    lat_error = abs(lat1 - lat2)
    lon_error = abs(lon1 - lon2)
    
    return lat_error, lon_error



# Calculate distances
df['Error_Best_Match_1_deg'] = df.apply(lambda x: calculate_lat_long_error(x['Location_coords'], x['Best_Match_1_coords']), axis=1)
df['Error_Best_Match_2_deg'] = df.apply(lambda x: calculate_lat_long_error(x['Location_coords'], x['Best_Match_2_coords']), axis=1)

df
```

```python
err_best_match_1_lat = df['Error_Best_Match_1_deg'].apply(lambda x: x[0]).mean()
err_best_match_1_lon = df['Error_Best_Match_1_deg'].apply(lambda x: x[1]).mean()
err_best_match_2_lat = df['Error_Best_Match_2_deg'].apply(lambda x: x[0]).mean()
err_best_match_2_lon = df['Error_Best_Match_2_deg'].apply(lambda x: x[1]).mean()

print("Error rate in lat and long for best match 1: {:.2f} deg, {:.2f} deg".format(err_best_match_1_lat, err_best_match_1_lon))
print("Error rate in lat and long for best match 2: {:.2f} deg, {:.2f} deg".format(err_best_match_2_lat, err_best_match_2_lon))
```

```python
# Extract latitude and longitude errors for best match 1
df['Error_Best_Match_1_Lat'] = df['Error_Best_Match_1_deg'].apply(lambda x: x[0])
df['Error_Best_Match_1_Lon'] = df['Error_Best_Match_1_deg'].apply(lambda x: x[1])

# Extract latitude and longitude errors for best match 2
df['Error_Best_Match_2_Lat'] = df['Error_Best_Match_2_deg'].apply(lambda x: x[0])
df['Error_Best_Match_2_Lon'] = df['Error_Best_Match_2_deg'].apply(lambda x: x[1])

# Plot histograms for latitude errors
plt.figure(figsize=(8, 8))

# Histogram for Best Match 1 Latitude Error
plt.subplot(3, 2, 1)
plt.hist(df['Error_Best_Match_1_Lat'].dropna(), bins=20, color='blue', alpha=0.7)
plt.title('Best Match 1 Latitude Error')
plt.xlabel('Degrees')
plt.ylabel('Frequency')

# Histogram for Best Match 1 Longitude Error
plt.subplot(3, 2, 2)
plt.hist(df['Error_Best_Match_1_Lon'].dropna(), bins=20, color='green', alpha=0.7)
plt.title('Best Match 1 Longitude Error')
plt.xlabel('Degrees')
plt.ylabel('Frequency')

# Histogram for Best Match 2 Latitude Error
plt.subplot(3, 2, 3)
plt.hist(df['Error_Best_Match_2_Lat'].dropna(), bins=20, color='red', alpha=0.7)
plt.title('Best Match 2 Latitude Error')
plt.xlabel('Degrees')
plt.ylabel('Frequency')

# Histogram for Best Match 2 Longitude Error
plt.subplot(3, 2, 4)
plt.hist(df['Error_Best_Match_2_Lon'].dropna(), bins=20, color='purple', alpha=0.7)
plt.title('Best Match 2 Longitude Error')
plt.xlabel('Degrees')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set the style of seaborn
sns.set_style("whitegrid")

# Plotting histograms/distribution plots for latitude and longitude errors using seaborn
plt.figure(figsize=(12, 10))

# Define a function to draw a line from the peak of the KDE to the x-axis
def draw_kde_peak_line(data, **kws):
    # Calculate the KDE
    density = sns.kdeplot(data).get_lines()[0].get_data()
    # Find the x value of the peak
    x_peak = density[0][np.argmax(density[1])]
    # Find the y value of the peak
    y_peak = np.max(density[1])
    # Draw a vertical line from the peak to the x-axis
    plt.vlines(x_peak, ymin=0, ymax=y_peak, colors='k', linestyles='dashed')

# Add subplots for each error type
for i, (col, color) in enumerate(zip(['Error_Best_Match_1_Lat', 'Error_Best_Match_1_Lon',
                                      'Error_Best_Match_2_Lat', 'Error_Best_Match_2_Lon'],
                                     ['red', 'red', 'purple', 'purple']), start=1):
    plt.subplot(2, 2, i)
    # Plot the histogram with KDE
    sns.histplot(df[col].dropna(), kde=True, color=color)
    # Draw the line from the peak of KDE to the x-axis
    draw_kde_peak_line(df[col].dropna())
    plt.title(f'{col.replace("_", " ")}')
    plt.xlabel('Error in Degrees')
    plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from math import pi

categories = ['Best Match 1', 'Best Match 2']
avg_lat_errors = [df['Error_Best_Match_1_Lat'].mean(), df['Error_Best_Match_2_Lat'].mean()]
avg_lon_errors = [df['Error_Best_Match_1_Lon'].mean(), df['Error_Best_Match_2_Lon'].mean()]

# We need to repeat the first value to close the circular graph:
avg_lat_errors += avg_lat_errors[:1]
avg_lon_errors += avg_lon_errors[:1]

# Calculate angle for each category
angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]

# Plot
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Draw one line per variable + fill
ax.fill(angles, avg_lat_errors, color='red', alpha=0.25)
ax.plot(angles, avg_lat_errors, color='red', linewidth=2, label='Latitude Error')

ax.fill(angles, avg_lon_errors, color='blue', alpha=0.25)
ax.plot(angles, avg_lon_errors, color='blue', linewidth=2, label='Longitude Error')

# Draw labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

# Add legend and title
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.title('Average Latitude and Longitude Errors for Matches and Predictions')

plt.show()
```
